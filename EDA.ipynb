```python
# Aggregate data by state
df_state = df.groupby('State').agg({
    'Rates.Marijuana.Used Past Month.18-25': 'mean',
    'Rates.Tobacco.Cigarette Past Month.18-25': 'mean',
    'Population.18-25': 'sum'
}).reset_index()

# Choose variables for x, y, and bubble size
x_variable = 'Rates.Marijuana.Used Past Month.18-25'
y_variable = 'Rates.Tobacco.Cigarette Past Month.18-25'
bubble_size_variable = 'Population.18-25'

# Plot the bubble chart
plt.figure(figsize=(14, 10))
plt.scatter(df_state[x_variable], df_state[y_variable], s=df_state[bubble_size_variable]/1000, alpha=0.5, c='skyblue', edgecolors='black')
plt.title('Bubble Chart: Marijuana Use vs. Tobacco Cigarette Use (Age 18-25)')
plt.xlabel('Marijuana Use Rate (Age 18-25)')
plt.ylabel('Tobacco Cigarette Use Rate (Age 18-25)')
plt.grid(True)

# Annotate points with state names
for i, txt in enumerate(df_state['State']):
    plt.annotate(txt, (df_state[x_variable][i], df_state[y_variable][i]), fontsize=8)

# Create custom legend without bubble
custom_legend = plt.Line2D([], [], linestyle='', marker='o', markersize=10, color='skyblue')
plt.legend([custom_legend], ['Population (Age 18-25)'])

plt.show()
```

```python
# Create a dictionary to store the top 5 states for each drug and age group
top_states_by_drug_age = {}

# Dictionary mapping drug names to column names for different age groups
drug_columns = {
    'Alcohol': [
        'Totals.Alcohol.Use Disorder Past Year.12-17',
        'Totals.Alcohol.Use Disorder Past Year.18-25',
        'Totals.Alcohol.Use Disorder Past Year.26+'
    ],
    'Tobacco': [
        'Totals.Tobacco.Cigarette Past Month.12-17',
        'Totals.Tobacco.Cigarette Past Month.18-25',
        'Totals.Tobacco.Cigarette Past Month.26+'
    ],
    'Cocaine': [
        'Totals.Illicit Drugs.Cocaine Used Past Year.12-17',
        'Totals.Illicit Drugs.Cocaine Used Past Year.18-25',
        'Totals.Illicit Drugs.Cocaine Used Past Year.26+'
    ],
    'Marijuana': [
        'Totals.Marijuana.Used Past Month.12-17',
        'Totals.Marijuana.Used Past Month.18-25',
        'Totals.Marijuana.Used Past Month.26+'
    ]
}

# Iterate over each drug and age group
for drug, columns in drug_columns.items():
    top_states_by_drug_age[drug] = {}
    for age_group, column in zip(['12-17', '18-25', '26+'], columns):
        # Calculate the total usage of the drug for each state and age group
        state_total = df.groupby('State')[column].sum()
        # Sort the states based on their usage of the drug for the current age group
        top_states = state_total.sort_values(ascending=False).head(5)
        # Add the top 5 states for the current age group to the dictionary
        top_states_by_drug_age[drug][age_group] = top_states

# Print the top 5 states for each drug and age group
for drug, age_groups in top_states_by_drug_age.items():
    print(f"Top 5 States for {drug}:")
    for age_group, states in age_groups.items():
        print(f"  Age Group {age_group}:")
        print(states)
    print()
```

```python
# Set the number of columns and rows for subplots
num_cols = 3  # Number of age groups
num_rows = len(drug_columns)  # Number of drugs

# Set the figure size
plt.figure(figsize=(18, 5*num_rows))

# Iterate over each drug
for i, (drug, columns) in enumerate(drug_columns.items()):
    # Iterate over each age group
    for j, column in enumerate(columns):
        # Calculate the total usage of the drug for each state and age group
        state_total = df.groupby('State')[column].sum()
        # Sort the states based on their usage of the drug for the current age group
        top_states = state_total.sort_values(ascending=False).head(5)

        # Create a subplot for the current drug and age group
        plt.subplot(num_rows, num_cols, i*num_cols + j + 1)
        # Plot the bar plot for the top 5 states
        top_states.plot(kind='bar', color='skyblue')
        # Set the title for the subplot
        plt.title(f'Top 5 States for {drug} (Age Group {columns[j].split(".")[-1]})')
        # Set the y-label for the subplot
        plt.ylabel('Total Usage')
        # Rotate x-labels for better readability
        plt.xticks(rotation=45)

# Adjust the layout to prevent overlap
plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Pivot the DataFrame to have years as columns and states as rows
pivot_df = df.pivot(index='State', columns='Year', values='Rates.Marijuana.New Users.18-25')

# Set the figure size
plt.figure(figsize=(12, 12))

# Create the heatmap
sns.heatmap(pivot_df, cmap='YlGnBu')

# Set the title and labels
plt.title('Rates of New Marijuana User (Age 18-25) Over Years')
plt.xlabel('Year')
plt.ylabel('State')

# Show the plot
plt.show()
```

```python
# Define the width of each bar
bar_width = 0.25

# Set the color palette
palette = sns.color_palette("husl", len(drug_columns))

# Iterate over each drug
for drug, columns in drug_columns.items():
    # Set the figure size
    plt.figure(figsize=(10, 6))

    # Define the index for the x-axis (years)
    index = np.arange(len(df['Year'].unique()))

    # Iterate over each age group
    for i, column in enumerate(columns):
        # Group data by year and calculate the total usage of the drug for each age group
        total_usage = df.groupby('Year')[column].sum()
        # Plot the grouped bar graph for the current age group
        plt.bar(index + i * bar_width, total_usage.values, bar_width, label=f'Age Group {column.split(".")[-1]}', color=palette[i])

    # Set labels, legend, and title for the plot
    plt.xlabel('Year')
    plt.ylabel('Total Usage')
    plt.title(f'Total Usage of {drug} by Age Group Over Years')
    plt.xticks(index + bar_width * (len(columns) - 1) / 2, df['Year'].unique())
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```
